```python
import random
```

```python
def number_guessing_game():
    print("Welcome to the Number Guessing Game!")
    random.randint(-100,100)
    secret_num = random.randint(-100,100)


    for i in range(1,11):
        try:
            user_input = int(input("Enter the guess number:"))
            if user_input < -100 or user_input > 100:
                print("Invalid Number!")
            elif user_input > secret_num:
                print("Number is too High!")
            elif user_input < secret_num:
                print("Number is too Low!")
            else:
                print(" Your Guessed number is Correct!")
                break
        
        except ValueError:
            print("Invalid Characters, Not Recognized!")

    if user_input == secret_num:
        print(f"Congratulations! You have Guessed the number {secret_num} in {i} attempts.")
    else:
        print(f"Oops! No turns Left. My Number was {secret_num}")
```

```python
number_guessing_game()
```
